```python
model = YOLO("yolov8s-obb.pt")
model
```

```python
results = model.train(
    data="/kaggle/input/obbcar-datayaml/data.yaml", 
    epochs=50, 
    imgsz=640, 
    amp=False,
    optimizer='Adam',
    plots=True,
    lr0=0.01,
    batch=4
)
```
